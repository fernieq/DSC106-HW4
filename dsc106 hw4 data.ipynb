```python
#jennifer lawrence leading roles movie data
j_law_data = pd.read_csv('/Users/fernieqin/Desktop/DSC106/hw4/j_law.csv')
j_law_data.dropna(inplace = True) 
j_law_data.head()
```

```python
# new data frame with split value columns 
year_month = j_law_data["Release\nDate"].str.split("," ,n = 1, expand = True) 
# making separate first month and date column from new data frame 
j_law_data["release_month_date"]= year_month[0] 
# #making separate last name column from new data frame 
j_law_data["release_year"]= year_month[1] 
#drop some columns
j_law_data.drop(columns =["Release\nDate"], inplace = True)
j_law_data.drop(columns =["Domestic\nShare"], inplace = True)
```

```python
#data preprocessing
j_law_data['Opening\nWeekend'] = j_law_data['Opening\nWeekend'].str.replace('$', '')
j_law_data['Opening\nWeekend'] = j_law_data['Opening\nWeekend'].str.replace(',', '')
j_law_data['Opening\nWeekend'] = j_law_data['Opening\nWeekend'].astype(int)
```

```python
j_law_data['Maximum\nTheaters'] = j_law_data['Maximum\nTheaters'].str.replace(',', '')
j_law_data['Maximum\nTheaters'] = j_law_data['Maximum\nTheaters'].astype(int)
```

```python
j_law_data['Domestic\nBox Office'] = j_law_data['Domestic\nBox Office'].str.replace('$', '')
j_law_data['Domestic\nBox Office'] = j_law_data['Domestic\nBox Office'].str.replace(',', '')
j_law_data['Domestic\nBox Office'] = j_law_data['Domestic\nBox Office'].astype(int)
```

```python
j_law_data['Worldwide\nBox Office'] = j_law_data['Worldwide\nBox Office'].str.replace('$', '')
j_law_data['Worldwide\nBox Office'] = j_law_data['Worldwide\nBox Office'].str.replace(',', '')
j_law_data['Worldwide\nBox Office'] = j_law_data['Worldwide\nBox Office'].astype(int)
```

```python
#rename columns
j_law_data = j_law_data.rename(columns={"Opening\nWeekend": "Opening_Weekend", "Maximum\nTheaters": "Maximum_Theaters",
                          "Domestic\nBox Office":"Domestic_Box Office", "Worldwide\nBox Office":"Worldwide_Box Office"})
```
